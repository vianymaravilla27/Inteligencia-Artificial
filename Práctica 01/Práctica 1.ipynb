2) Separe el dataset en un conjunto de entrenamiento (60%) y de prueba (40%) sin mezclar datos

3) Guarde los siguientes archivos

  - x_e.csv (datos del conjunto de entrenamiento)
  
  - t_e.csv (etiquetas del conjunto de entrenamiento)

```python
# Import the libraries necessary for the realization of the practice.
# Importar las librerias necesarias para la realización de la practica.

import pandas as pnd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Configure pandas dataframes for visualization 
# Configurar los dataframes de pandas para su visualización 

pnd.set_option('display.max_rows', None)
pnd.set_option('display.max_columns', None)
pnd.set_option('display.width', 1000)
pnd.set_option('display.colheader_justify', 'center')
pnd.set_option('display.precision', 3)
```

```python
# From the file "heart" we create a dataframe
# A partir del archivo "heart" creamos un dataframe 

dataf = pnd.read_csv('heart.csv')

# Visualization of the dataframe 
# Visualización del dataframe 
dataf
```

```python
# Assigning the name of the label column
# Asignamos el nombre de la columa de la etiqueta

name_tg = 'target'
```

```python
# Corpus without tags
# Corpus sin etiquetas
# X

corpus_stg = dataf.drop(name_tg, axis = 1).values
```

```python
# Visualization
# Visualización

corpus_stg
```

```python
# With tags
# Con etiquetas
# Y
corpus_tag = dataf[name_tg].values
```

```python
# Visualization 
# Visualización 

corpus_tag
```

```python
# Training and test dataset 
# Conjuto de datos de entrenamiento y prueba 
x_e, y_e, x_p, y_p =  train_test_split(corpus_stg, corpus_tag, test_size = 0.4, shuffle = False)  
```

```python
# To be able to save CSV files we can first extract the name of the columns
# Para poder guardar los archivos CSV podemos primero extraer el nombre de las columnas

names_c = list(dataf.columns.values)
names_s = ','.join([column for column in names_c if column != name_tg])

# Columns names
# Nombre de las columnas

print(names_s)
```

```python
# Saving of the files x_e, y_e, x_p, y_p, in the desired path
# Guardado de los archivos x_e, y_e, x_p, y_p, en la ruta deseada

# With numpy 
# Con numpy

np.savetxt("./data/x_e.csv", x_e, delimiter=",", fmt="%d",
           header= names_s, comments="")

np.savetxt("./data/y_e.csv", y_e, delimiter=",", fmt="%d",
           header= name_tg, comments="")

np.savetxt("./data/x_p.csv", x_p, delimiter=",", fmt="%d",
           header= names_s, comments="")

np.savetxt("./data/y_p.csv", y_p, delimiter=",", fmt="%d",
           header= name_tg, comments="")
```
